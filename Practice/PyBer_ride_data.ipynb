```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
pyber_raw_df = pd.read_csv('Resources/PyBer_ride_data.csv')
pyber_raw_df
```

```python
#This cell does mostly the same thing as the next cell, only difference is that in that pyber.plot only
#gives some of the months on the x_axis, next cell uses tick locations to remedy this situation.

#pyber_raw_df.plot(pyber_raw_df('Month'),pyber_raw_df('Avg. Fare ($USD)') )
plt.figure(figsize = (10,10))
Months = pyber_raw_df['Month']
Fare = pyber_raw_df['Avg. Fare ($USD)']
plt.plot(Months, Fare)

#pyber_raw_df.plot(x='Month',y='Avg. Fare ($USD)')
```

```python
x_axis = np.arange(len(pyber_raw_df))
tick_locations = [value for value in x_axis]
pyber_raw_df.plot(x='Month',y='Avg. Fare ($USD)')
plt.xticks(tick_locations, pyber_raw_df['Month'])
#plt.show()
```

```python
import statistics as stats
stand_dev = stats.stdev(pyber_raw_df['Avg. Fare ($USD)'])
pyber_raw_df.plot.bar(x='Month', y='Avg. Fare ($USD)', yerr = stand_dev, capsize = 3, color = 'skyblue')
#plt.xaxis()

#https://stackoverflow.com/questions/10998621/rotate-axis-text-in-python-matplotlib
plt.xticks(rotation=0)
plt.yticks(np.arange(0,61,step=5))
#plt.xticks(rotation=90)
plt.show()
```
